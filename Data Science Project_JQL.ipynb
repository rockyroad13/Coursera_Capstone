```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table',{'class':'wikitable sortable'}).tbody

rows = table.find_all('tr')

columns = [v.text.replace('\n', '') for v in rows[0].find_all('th')]

df = pd.DataFrame(columns=columns)

for i in range(1, len(rows)):
    tds = rows[i].find_all('td')

    if len(tds) == 3:
        values = [tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[2].text.replace('\n', '')]
    else:
        values = [td.text.replace('\n', '') for td in tds]

    df = df.append(pd.Series(values, index=columns), ignore_index=True)
    df.to_csv('pc.csv')
    
df1 = df[~df['Borough'].isin(['Not assigned'])]
df1.reset_index(drop=True, inplace=True)

for index, row in df1.iterrows():
    if row['Neighborhood'] == 'Not assigned':
        row['Neighborhood'] == row['Borough']
df1.to_csv('pc1.csv')
df1.head()
```

## Looking for a good place to open a Gym in Toronto Area

```python
# Select only boroughs that contain the word "Toronto" in Toronto data.
toronto_t = toronto_data.loc[toronto_data['Borough'].str.contains('Toronto')]
toronto_t.reset_index(drop=True, inplace=True)
toronto_t
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
                  
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
                  
        results = requests.get(url).json()["response"]['groups'][0]['items']
            
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_t['Neighborhood'],
                                   latitudes=toronto_t['Latitude'],
                                   longitudes=toronto_t['Longitude']
                                  )
```

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix=" ", prefix_sep="")
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 


columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[columns]
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby(["Neighborhood"]).mean().reset_index()
toronto_grouped
```

```python
len(toronto_grouped[toronto_grouped[' Gym'] > 0])
```

```python
gym_t=toronto_grouped[['Neighborhood',' Gym']]
gym_t.head(10)
```

```python
nclusters = 3
nclustering = gym_t.drop(["Neighborhood"], 1)

kmeans = KMeans(n_clusters=nclusters, random_state=1)
kmeans.fit_transform(nclustering)

kmeans.labels_[0:10]
```

```python
g_merged = gym_t.copy()

# add clustering labels
g_merged["Cluster Labels"] = kmeans.labels_
```

```python
g_merged
```

```python
g_merged = g_merged.join(toronto_venues.set_index("Neighborhood"), on="Neighborhood")
print(g_merged.shape)
g_merged.head()
```

```python
g_merged.sort_values(["Cluster Labels"], inplace=True)
g_merged.head()
```

```python
map_clusters = folium.Map(location=[latitude, longitude],zoom_start=11)

m_colors={}
m_colors[0] = 'red'
m_colors[1] = 'yellow'
m_colors[2] = 'blue'



for lat, lon, cluster in zip(g_merged['Neighborhood Latitude'], g_merged['Neighborhood Longitude'], g_merged['Cluster Labels']):
    
    
    folium.features.CircleMarker(
        [lat, lon],
        radius=5, 
        color =m_colors[cluster],
        fill_color=m_colors[cluster],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
